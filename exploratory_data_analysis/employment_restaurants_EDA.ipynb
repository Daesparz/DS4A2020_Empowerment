```python
import pandas as pd
import datetime as date
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = '../data/raw/economics/Employment/Unemployment_Restaurants/SeriesReport-20201211121725_1bd446.xlsx'

df = pd.read_excel(path, index_col=0, header=12)  
```

```python
df.head()
```

```python
df_employment = df.reset_index().melt(id_vars='Year', var_name='Month', value_name='Employees')
```

```python
df_employment['Date'] = pd.to_datetime(df_employment['Year'].astype(str)  + df_employment['Month'], format='%Y%b')
```

```python
sns.set_style("whitegrid")

fig, ax = plt.subplots(figsize=(10.5, 4.5))
_ = sns.lineplot(data=df_employment, x='Date', y='Employees')
_ = plt.xlabel('Date')
_ = plt.ylabel('Employees (Thousand)')
_ = plt.title('Employment from 1990 to 2020 in the Food Services and Drinking Places Industry')
#_ = plt.xscale('log')
```

```python
sns.set_style("whitegrid")

fig, ax = plt.subplots(figsize=(10.5, 4.5))
_ = sns.lineplot(data=df_employment[df_employment.Year >=2020], x='Date', y='Employees')
_ = plt.xlabel('Date')
_ = plt.ylabel('Employees (Thousand)')
_ = plt.title('Monthly employment during 2020 in the Food Services and Drinking Places Industry')
```
